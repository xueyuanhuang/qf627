### Timing

```python
import time

def Timing(Time):
    
    while (Time):
        minutes, seconds = divmod(Time, 60)
        print("{:02d}:{:02d}".format(minutes, seconds), end = "\r")
        time.sleep(1)
        Time -= 1
```

```python
Timing(100)
```

### Question 6. Please add `City` column: Create the `City` from `Purchase Address` column

```python
A6 = A5.copy()
```

```python
A6["Purchase Address"] = A6["Purchase Address"].astype("string")
```

```python
def city(purchase_address):
    return purchase_address.split(",")[1].strip()
def state(purchase_address):
    return purchase_address.split(",")[2].split(" ")[1]

A6["City_State"] = A6["Purchase Address"].apply(lambda x: f"{city(x)}_{state(x)}")
```
