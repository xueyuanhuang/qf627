## Who are the top 10% highest-paid players?

> Let's calculate the 90 percentile cutoff.

```python
data_CLEANED.hist("yearID", bins = 80)
```

> Let's filter out players that make more money than the cutoff.

```python
cutoff = data_CLEANED.salary.quantile(q = .90)
cutoff
```

> Let's use the `nlargest` to see the top 10 best paid players.

```python
top10_percent_wage_earners = data_CLEANED[data_CLEANED["salary"] >= cutoff]
```

```python
top10_percent_wage_earners[["playerID"]]
```

```python
TOP10_players=top10_percent_wage_earners.nlargest(10, "salary")
```

```python
TOP10_players[["playerID", "salary"]]
```

```python
top10_percent_wage_earners.nsmallest(10, "salary").playerID
```

> Now let's visualize the results.

```python
TOP10_players.plot(kind = "bar", x = "nameGiven", y = "salary")
```
