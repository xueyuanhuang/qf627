### Random Assignment for Take-Home Assignment Task

```python
import random
import pandas as pd
```

```python
roster = pd.read_csv("qf627_g1_for_project.csv")
roster
```

```python
students = roster["name"].tolist()
```

```python
random.seed(20231027)
```

```python
members_per_group = 5
```

```python
group = 1

for s in students[:]:
    if members_per_group == 5:
        print("***** Group {} *****".format(group)
             ) 
        members_per_group = 0
        group += 1
    
    student = random.choice(students)
    
    print(student)
    
    members_per_group += 1
    
    students.remove(str(student)
                   )
```
